### Za≈Çadowanie bibliotek i danych


```python
import numpy as np
import pandas as pd

def fetch_financial_data(company='AMZN'):
    """
    This function fetches stock market quotations.
    """
    import pandas_datareader.data as web
    return web.DataReader(name=company, data_source='stooq')

df = fetch_financial_data('META')
df.head()
```

```python
df['2019-12-04': '2019-12-10']
```

```python
df.to_csv('fb.csv') # csv = comma sprited value
```

```python
df_nov = df[(df.index.month == 11) & (df.index.year == 2019)]
df_nov
```

```python
df_nov.to_csv('fb_nov.csv') # to_csv - zapisuje dane w pliku .csv
```

```python
new_df = pd.read_csv('fb_nov.csv') # pd.read_<rozszerzenie> - odczytuje dane z pliku
new_df
```
